```python
q1 = new_data["registered"].quantile(.25)
q3 = new_data["registered"].quantile(.75)
```

```python
def find_outliers(df, cols):
    q1 = df[cols].quantile(.25)
    q2 = df[cols].quantile(.75)
    iqr = q3 - q1 
    min_r = q1 - 1.5*iq
    max_r = q3 + 1.5*iqr
    outliers_indices = df.index[(df[cols] < min_r ) | (df[cols] > max_r)]
    return outliers_indices
```

```python
find_outliers(new_data, "registered")
```
